```python
import agate
import csv
```

```python
voci = agate.Table.from_csv('/Users/gu/Workspace/open-bilanci-armonizzati/resources/data/pcox_voci.csv')
voci_corrette = []
```

```python
for n, voce in enumerate(voci):
    voce_corretta = [x for x in voce]
    if voce[2] in (4,5,6) and voce[-1] == '- previsione di cassa':
        voce_corretta[-1] = voci[n-3][-1].replace('competenza', 'cassa')
    voci_corrette.append(voce_corretta)
```

```python
voci_corrette_table = agate.Table(
    voci_corrette,
    voci.column_names,
    voci.column_types
)
```

```python
voci_corrette_table.to_csv('/Users/gu/Workspace/open-bilanci-armonizzati/resources/data/pcox_voci_corrette.csv')
```
